```python
from googleapiclient.discovery import build
import pandas
import emoji
import re
import warnings

warnings.filterwarnings('ignore')
api_key = "AIzaSyCgJDfT5eyRi6M9u893npJRC7gQ6Ql_YQU"

def get_comments_from_channel(video_id_list, file_name):

    df = pandas.DataFrame(columns=['text', 'emoji'])

    resource = build('youtube', 'v3', developerKey=api_key)
    len_ids_lst = len(video_id_list)
    for i in range(len_ids_lst):
        print(str(i + 1) + " out of " + str(len_ids_lst))
        try:
            request = resource.commentThreads().list(
                part="snippet",
                videoId=video_id_list[i],
                maxResults=10000,
                order="orderUnspecified")

            response = request.execute()

            items = response["items"]

            for item in items:

                text = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
                str_free_emoji = re.sub(emoji.get_emoji_regexp(), r'', text)
                if not is_heb(str_free_emoji):
                    continue
                
                emojis = re.findall(emoji.get_emoji_regexp(), text)
                if (len(emojis)!=0) and (len(str_free_emoji)!=0):
                    emojis_set = set(emojis)
                    for emoji_i in set(emojis):
                        df = df.append({'text':str(str_free_emoji), 'emoji':emoji_i}, ignore_index=True)
        except:
            print("Caught exception")
    file_name = file_name + '.csv'
    df.to_csv(file_name, mode='a', index=False, header=False)

```

```python
# עלול לפעמים לטעות כי זה בודק רק לפי האות הראשונה

import unicodedata
import string
def is_heb(text):
    text = text.translate(str.maketrans('', '', string.punctuation))
    return 'HEBREW' in unicodedata.name(text.strip()[0])
```

```python
from googleapiclient.discovery import build
import pandas
import emoji
import re

def get_channel_videos(channel_id):
    
    youtube = build('youtube', 'v3', developerKey=api_key)
    
    res = youtube.channels().list(id=channel_id, 
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id, 
                                           part='snippet', 
                                           maxResults=5000,
                                           pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
    
    videos_id_list = []
    for video in videos:
        videos_id_list.append(video['snippet']['resourceId']["videoId"])
        
    return videos_id_list
```
